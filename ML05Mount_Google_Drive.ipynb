```python
# ------------------------------
# STEP 1: Install dependencies
# ------------------------------
!pip install opencv-python-headless mtcnn

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from mtcnn import MTCNN

# ------------------------------
# STEP 2: Load Pretrained Models
# ------------------------------
# Age & Gender Models
age_proto = "https://raw.githubusercontent.com/spmallick/learnopencv/master/AgeGender/AgeGender/age_deploy.prototxt"
age_model = "https://data.vision.ee.ethz.ch/cvl/rrothe/imdb-wiki/static/age_net.caffemodel"

gender_proto = "https://raw.githubusercontent.com/spmallick/learnopencv/master/AgeGender/AgeGender/gender_deploy.prototxt"
gender_model = "https://data.vision.ee.ethz.ch/cvl/rrothe/imdb-wiki/static/gender_net.caffemodel"

!wget -q $age_proto -O age_deploy.prototxt
!wget -q $age_model -O age_net.caffemodel
!wget -q $gender_proto -O gender_deploy.prototxt
!wget -q $gender_model -O gender_net.caffemodel

# ------------------------------
# STEP 3: Load Networks and Face Detector
# ------------------------------
age_net = cv2.dnn.readNet("age_net.caffemodel", "age_deploy.prototxt")
gender_net = cv2.dnn.readNet("gender_net.caffemodel", "gender_deploy.prototxt")
detector = MTCNN()

# Labels
AGE_BUCKETS = ['(0-2)', '(4-6)', '(8-12)', '(15-20)', '(25-32)', '(38-43)', '(48-53)', '(60-100)']
GENDER_LIST = ['Male', 'Female']

# ------------------------------
# STEP 4: Upload and Detect
# ------------------------------
uploaded = files.upload()
for fn in uploaded.keys():
    img = cv2.imread(fn)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # MTCNN expects RGB

    # Detect faces
    faces = detector.detect_faces(img_rgb)

    if faces:
        for face in faces:
            x, y, w, h = face['box']
            # Extract face
            face_img = img[y:y+h, x:x+w]
            face_blob = cv2.dnn.blobFromImage(face_img, 1.0, (227, 227), (78.4263377603, 87.7689143744, 114.895847746), swapRB=False)

            # Predict Gender
            gender_net.setInput(face_blob)
            gender_preds = gender_net.forward()
            gender = GENDER_LIST[gender_preds[0].argmax()]
            gender_conf = gender_preds[0].max()

            # Predict Age
            age_net.setInput(face_blob)
            age_preds = age_net.forward()
            age = AGE_BUCKETS[age_preds[0].argmax()]
            age_conf = age_preds[0].max()

            # Draw bounding box and text
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            overlay_text = f"{gender} ({gender_conf*100:.2f}%), {age} ({age_conf*100:.2f}%)"
            cv2.putText(img, overlay_text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Display image with detections
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()
    else:
        print("No faces detected.")
```
